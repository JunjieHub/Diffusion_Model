```python
%cd CO2_plume_reconstruction
from utility import get_penetrate_index

receivers_table = [] 
intervals = np.linspace(3,8,6)
for inv in intervals:
    receivers = []
    x = 0
    while x<68:
        receivers.append([int(x), 20])
        x += inv
    receivers_table.append(receivers)

senders = [[33,20]]
fig, axes = plt.subplots(2,3)
fig.set_size_inches((15,4))
axes = axes.ravel()
for i in range(6):
    receivers = receivers_table[i]
    grid = np.zeros((20,68))
    p_index = get_penetrate_index(senders, receivers)
    for x,y in p_index:
        grid[y][x] = 1
    axes[i].imshow(grid, origin = 'lower', cmap = 'jet')
    axes[i].plot(senders[0][0],senders[0][1], 'ro', ms = 10)
    for r in receivers:
        axes[i].plot(r[0],r[1], 'go', ms = 5)
        

fig.savefig('geophone_config1.png', dpi = 300, bbox_inches = 'tight')
```
